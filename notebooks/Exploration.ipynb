```python
def plot_ts(contact_type, start_year, curr_year):
    plt.figure(figsize=(15, 10))
    
    prev_year = curr_year - 1
    
    # plot JAN - MAR for all years including current year.
    for i in range(start_year, curr_year+1):
        mask = (contacts_df.dates.dt.month.isin([1, 2, 3])) & (contacts_df.dates.dt.year == i)

        ss = contacts_df.loc[mask & (contacts_df.contact_types == contact_type), :]\
               .sort_values(by='dates')\
               ['num_contacts'].values
        
#         ss = ss.assign(day=ss.dates.dt.day)
#         ss = ss.groupby(['weekday'])['num_contacts'].mean()

        plt.scatter(np.arange(len(ss)), ss, label='%s'%(i))
    
    
    # prev year OCT - DEC num contacts
    mask = (contacts_df.dates.dt.month.isin([12])) & (contacts_df.dates.dt.year.isin([prev_year]))

    ss = contacts_df.loc[mask & (contacts_df.contact_types == contact_type), :]\
               .sort_values(by='dates')\
               ['num_contacts'].values
            
#     ss = ss.assign(day=ss.dates.dt.day)
#     ss = ss.groupby(['weekday'])['num_contacts'].mean()
    
    plt.scatter(np.arange(len(ss)), ss, label='%s_END'%(prev_year))
    plt.legend(loc='best');
```

** Overlap of num contacts for a contact (2014, 2015, 2016) between January and March **

```python
def plot_overlap(contact_type):

    # 2014
    mask = (contacts_df.dates.dt.month.isin([1, 2, 3])) & (contacts_df.dates.dt.year == 2014)

    _2014_call = contacts_df.loc[mask & (contacts_df.contact_types == contact_type), :]\
                 .sort_values(by='dates')\
                 ['num_contacts'].values

    plt.scatter(np.arange(len(_2014_call)), _2014_call, c='r', alpha=.3, label='2014')

    # 2015
    mask = (contacts_df.dates.dt.month.isin([1, 2, 3])) & (contacts_df.dates.dt.year == 2015)

    _2015_call = contacts_df.loc[mask & (contacts_df.contact_types == contact_type), :]\
               .sort_values(by='dates')\
               ['num_contacts'].values

    plt.scatter(np.arange(len(_2015_call)), _2015_call, c='g', alpha=.4, label='2015')

    # 2016
    mask = (contacts_df.dates.dt.month.isin([1, 2, 3])) & (contacts_df.dates.dt.year == 2016)

    _2016_call = contacts_df.loc[mask & (contacts_df.contact_types == contact_type), :]\
                 .sort_values(by='dates')\
                 ['num_contacts'].values

    plt.scatter(np.arange(len(_2016_call)), _2016_call, c='b', alpha=.5, label='2016')
    plt.legend(loc='best')
    plt.title(contact_type);
```

** Calculate Averages **

```python
mask = contacts_df.dates >= '2014-01-01'

def get_quarter_mean(row):
    prev_quarter_start_date = row['dates'] + pd.DateOffset(-120)
    
    mask = (contacts_df.dates > prev_quarter_start_date.values[0]) & (contacts_df.dates < row['dates'].max()) \
           & (contacts_df.contact_types == row['contact_types'].values[0])
    
    return contacts_df.loc[mask, 'num_contacts'].mean()

def get_last_half_year_mean(row):
    prev_start_date = row['dates'] + pd.DateOffset(-182)
    
    mask = (contacts_df.dates > prev_start_date.values[0]) & (contacts_df.dates < row['dates'].max()) \
           & (contacts_df.contact_types == row['contact_types'].values[0])
        
    return contacts_df.loc[mask, 'num_contacts'].mean()

def get_last_year_mean(row):
    prev_start_date = row['dates'] + pd.DateOffset(-365)
    
    mask = (contacts_df.dates > prev_start_date.values[0]) & (contacts_df.dates < row['dates'].max()) \
           & (contacts_df.contact_types == row['contact_types'].values[0])
        
    return contacts_df.loc[mask, 'num_contacts'].mean()

def get_last_2_year_mean(row):
    prev_start_date = row['dates'] + pd.DateOffset(-365 * 2)
    
    mask = (contacts_df.dates > prev_start_date.values[0]) & (contacts_df.dates < row['dates'].max()) \
           & (contacts_df.contact_types == row['contact_types'].values[0])
        
    return contacts_df.loc[mask, 'num_contacts'].mean()

prev_quarter_mean   = contacts_df.loc[mask, :].groupby(['dates', 'contact_types']).apply(get_quarter_mean)
last_half_year_mean = contacts_df.loc[mask, :].groupby(['dates', 'contact_types']).apply(get_last_half_year_mean)
last_year_mean      = contacts_df.loc[mask, :].groupby(['dates', 'contact_types']).apply(get_last_year_mean)
last_2_year_mean    = contacts_df.loc[mask, :].groupby(['dates', 'contact_types']).apply(get_last_2_year_mean)
```

```python
quarter_mean     = contacts_df.loc[mask , :][['dates', 'contact_types']].apply(lambda x: prev_quarter_mean[x[0], x[1]], axis=1)
half_year_mean   = contacts_df.loc[mask , :][['dates', 'contact_types']].apply(lambda x: last_half_year_mean[x[0], x[1]], axis=1)
full_year_mean   = contacts_df.loc[mask , :][['dates', 'contact_types']].apply(lambda x: last_year_mean[x[0], x[1]], axis=1)
_2_year_mean     = contacts_df.loc[mask , :][['dates', 'contact_types']].apply(lambda x: last_2_year_mean[x[0], x[1]], axis=1)
```

```python
def plot_ts_res(category, start_year, curr_year):
    plt.figure(figsize=(15, 10))
    
    prev_year = curr_year - 1
    
    # plot JAN - MAR for all years including current year.
    for i in range(start_year, curr_year+1):
        mask = (resolutions_df.dates.dt.month.isin([1, 2, 3])) & (resolutions_df.dates.dt.year == i)

        ss = resolutions_df.loc[mask & (resolutions_df.categories == category), :]\
               .sort_values(by='dates')\
               ['num_resolutions'].values
        
#         ss = ss.assign(day=ss.dates.dt.day)
#         ss = ss.groupby(['weekday'])['num_contacts'].mean()

        plt.scatter(np.arange(len(ss)), ss, label='%s'%(i))
    
    
    # prev year OCT - DEC num contacts
    mask = (resolutions_df.dates.dt.month.isin([12])) & (resolutions_df.dates.dt.year.isin([prev_year]))

    ss = resolutions_df.loc[mask & (resolutions_df.categories == category), :]\
               .sort_values(by='dates')\
               ['num_resolutions'].values
            
#     ss = ss.assign(day=ss.dates.dt.day)
#     ss = ss.groupby(['weekday'])['num_contacts'].mean()
    
    plt.scatter(np.arange(len(ss)), ss, label='%s_END'%(prev_year))
    plt.legend(loc='best');
```
